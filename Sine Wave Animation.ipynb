```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

from IPython import display
```

```python
# Initialize figure object
fig = plt.figure()

# Mark the x-axis and y-axis
axis = plt.axes(xlim=(0,4),
               ylim=(-2,2))

# Initialize a line variable
line, = axis.plot([], [], lw=3)
```

```python
x=np.linspace(0, 4, 1000)

# Data which the line will contain (x,y)
def init():
    line.set_data([], [])
    return line,

def animate(i):  
    # Plot a sine wave
    y = np.sin(2*np.pi*(x-0.01*i))
    line.set_data(x,y)
    
    return line,

# anim = FuncAnimation(fig, animate,
#                     init_func = init,
#                     frames = 200,
#                     interval = 20,
#                     blit = True)
# anim = FuncAnimation(plt.gcf(), animate, init_func=init, interval=1000, frames = 500, repeat = False)
anim = FuncAnimation(fig, animate, init_func=init, frames = 250, interval=33, repeat = False)

# keyword arguments:

# repeat = False, terminates the animation after the specified number of frames.

# interval, the number of milliseconds between animation frames (default is 200). After each call to `animate`, 
# `FuncAnimation` waits the prescribed number of milliseconds before making the next call.

# frames, the total number of animation frames, which controls how many times `FuncAnimation` calls `animate.`

video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
# plt.tight_layout()
# plt.show()
# anim.save('continuousSineWave.mp4', writer='Pillow', fps=30)
```
